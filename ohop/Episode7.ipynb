````python
import ast

class RewriteName(ast.NodeTransformer):

    def visit_Name(self, node):
        return ast.Subscript(
            value=ast.Name(id='data', ctx=ast.Load()),
            slice=ast.Constant(value=node.id),
            ctx=node.ctx
        )
````

````python
rewriter = RewriteName()

def run_rewriter(cell, shell):
    return rewriter.visit(ast.parse(cell, '<transformer-metamagic-cell>'))
````

````python
%%mycell run_rewriter
def oof(rab):
    return rab ** 2 + rab + 1
````

````python
class MyVisitor(ast.NodeVisitor):
    def __init__(self):
        self.data = set()
        
    def visit_Constant(self, node):
        self.data.add(node)

class MyTransformer(ast.NodeTransformer):
    def __init__(self):
        self.visitor = MyVisitor()
        
    def visit_BinOp(self, node):
        if isinstance(node.op, ast.Add):
            left = self.visit(node.left)
            self.visitor.visit(left)
            right = self.visit(node.right)
            self.visitor.visit(right)
            if left in self.visitor.data and right in self.visitor.data:
                return ast.Constant(left.value + right.value)
            else:
                return ast.BinOp(left, node.op, right)
        return self.generic_visit(node)

def add_folder(cell):
    return ast.fix_missing_locations(
        MyTransformer().visit(ast.parse(cell))
    )

def just_parse(cell):
    return ast.parse(cell)
````
