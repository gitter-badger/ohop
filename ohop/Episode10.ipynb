# Name Substitution

Recognize `ast.Name()` nodes, replacing any matches with an AST expression... 

```python
class NameSubstitutionTransformer(ast.NodeTransformer):
    def __init__(self, **substitutions):
        self.substitutions = substitutions
        
    def visit_Name(self, node):
        if node.id in self.substitutions and isinstance(node.ctx, ast.Load):
            substitution = self.substitutions[node.id]
            print(ast.dump(node, indent=2))
            print('-->')
            print(ast.dump(substitution, indent=2))
            return substitution
        return self.generic_visit(node)

transformer = NameSubstitutionTransformer(x=ast.Constant(value=42))
target_code = '''
def wonky(x):
    return x ** 2 + x + 1
'''
target_ast = ast.parse(target_code)
result_ast = ast.fix_missing_locations(transformer.visit(target_ast))
print(ast.unparse(result_ast))
```

# Parametric Substitution

Recognize `ast.Call()` nodes, replacing any matches with the result of a function call...

```python
class ParametricSubstitutionTransformer(ast.NodeTransformer):
    def __init__(self, **substitutions):
        self.substitutions = substitutions
        
    def visit_Call(self, node):
        func = node.func
        fire = (
            isinstance(func, ast.Name) and
            func.id in self.substitutions and
            isinstance(func.ctx, ast.Load) and
            not node.keywords
        )
        if fire:
            substitution = self.substitutions[func.id](*node.args)
            print(ast.dump(node, indent=2))
            print('-->')
            print(ast.dump(substitution, indent=2))
            print('===')
            return substitution
        return self.generic_visit(node)

def const42(*args):
    return ast.Constant(value=42)

transformer2 = ParametricSubstitutionTransformer(x=const42)

target_code2 = '''
def wonky2(x):
    return x() ** 2 + x(999) + 1
'''

target_ast2 = ast.parse(target_code2)
print(ast.unparse(target_ast2))
print('==>')
result_ast2 = ast.fix_missing_locations(transformer2.visit(target_ast2))
print(ast.unparse(result_ast2))
```

```python
def poly_expr(x=ast.Constant(value=42)):
    return ast.BinOp(
        left=ast.BinOp(
              left=ast.BinOp(
                left=x,
                op=ast.Pow(),
                right=ast.Constant(value=2)),
              op=ast.Add(),
              right=x
        ),
        op=ast.Add(),
        right=ast.Constant(value=1)
    )

transformer3 = ParametricSubstitutionTransformer(x=poly_expr)
target_ast2 = ast.parse(target_code2)
print(ast.unparse(target_ast2))
print('==>')

result_ast3 = ast.fix_missing_locations(transformer3.visit(target_ast2))
print(ast.unparse(result_ast3))
```
