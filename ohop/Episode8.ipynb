```python
expr_module = ast.parse('1 + 2 + zab + 3 + 4')
print(expr_module)
```

```python
class AddChainer(ast.NodeVisitor):
    def __init__(self, *args, **kws):
        self.add_chains: typing.Mapping[ast.expr, typing.List] = {}

    def visit_BinOp(self, node):
        self.generic_visit(node)
        if isinstance(node.op, ast.Add):
            left_chain = self.add_chains.get(node.left, [node.left])
            right_chain = self.add_chains.get(node.right, [node.right])
            self.add_chains[node] = left_chain + right_chain

chainer = AddChainer()
chainer.visit(expr_module)
print(chainer.add_chains)
```

```python
class ConstantFolder2(ast.NodeTransformer):
    def visit_BinOp(self, node):
        if isinstance(node.op, ast.Add):
            left = self.visit(node.left)
            right = self.visit(node.right)
            if isinstance(left, ast.Constant) and isinstance(right, ast.Constant):
                result = ast.Constant(value=left.value + right.value)
            else:
                result = ast.BinOp(left=left, op=ast.Add(), right=right)
            return result
        return self.generic_visit(node)

demo = ast.fix_missing_locations(ConstantFolder2().visit(expr_module))
print(ast.unparse(demo))
```
